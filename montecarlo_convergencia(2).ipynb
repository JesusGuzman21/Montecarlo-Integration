```python
import matplotlib.pyplot as plt
import math
import numpy as np
from numpy.random import rand
def func(x):
    return np.exp(-x**2) # función a integrar

area_real = np.sqrt(np.pi)*math.erf(1)
print("área real: ",area_real)


def montecarlo_convergencia(f):
    a = -1
    b = 1
    j = 0 #contador
    er = 1
    s = 0.01 # error máximo
    while er > s:
        
        
        m = 0
        ys = f(np.linspace(a,b,100)) # 100 ys aleatorias
        lim_y = np.max(ys)
        vector_N = [0]
        vector_error = []
        N = 100

        for i in range(1,N+1):
            
            x = a + (b-a)*rand()
            y = rand()*lim_y
            fx = f(x)
    
            if y <= fx:
                m = m+1
        
    
    
        area = (b-a)* (m/N)
    
        er = abs(area-area_real) # no deja usar palabra error
        vector_error.append(er)

        N = N*10
        vector_N.append(N)
        j = j+1


    print(N)
    print(area)
convergencia=montecarlo_convergencia(func)
convergencia
```
