```python
import math
import numpy as np
from numpy.random import rand
from matplotlib.pyplot import  *
def func(x):
    return np.exp(-x**2) # función a integrar




def montecarlo(f):
    a = -1
    b = 1  #limites x
    m = 0
    N = 100000  # iteraciones
    ys = f(np.linspace(a,b)) # 100 ys aleatorias
    lim_y = max(ys)  # limite y
    for i in range(1,N+1):
        x = a + (b-a)*rand()
        y = rand() * lim_y
        fx = f(x)
        if y <= fx:
            m = m+1
    area_real = np.sqrt(np.pi)*math.erf(1)
    print("área real: ",area_real)
    area = (b-a)* (m/N)
    print("Área calculada: ",area)
    error = abs(area-area_real)
    print("Error: ",error)


```

```python
integral=montecarlo(func)
integral
```
